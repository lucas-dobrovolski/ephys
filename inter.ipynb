```python

import h5py
import numpy as np
import matplotlib.pyplot as plt

h5_path = "dataset.h5"

f = h5py.File(h5_path, "r")

f.visititems(lambda name, obj: print(name, obj))
```

```python
print(f["events"][2])
```

```python

high = 40000

plt.Figure(figsize=(15,6))

plt.plot(f["time"][0,:high],f["volt"][0,:high])
```

```python

# Elegir un dataset para inspeccionar
dataset_key = "time"  # cambi esto por el nombre que quieras

# Cargar los datos
data = f[dataset_key][0, -10:-1]
```

```python
data.shape
```

```python


# Ver shape y tipo
print(data.shape, data.dtype)

# Mostrar primeros valores
print(data[:10])

# Graficar si es 1D
if data.ndim == 1:
    plt.plot(data[:1000])
    plt.show()
# Graficar si es 2D
elif data.ndim == 2:
    plt.imshow(data[:100, :100], aspect="auto", cmap="viridis")
    plt.colorbar()
    plt.show()
```
